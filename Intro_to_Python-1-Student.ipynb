## Dictionaries (and DataFrames)

The dictionary is another datatype. It can be used in more complicated systems such as dataframes. Dictionaries are used to store multiple variables for a specific "key". The "keys" are just strings that we can use to index with instead of numbers! You can think of a dataframe just like a table in an excel spreadsheet. It is useful to use these for datasets with multiple columns to ensure that you are always using the correct information as it is much easier to forget what index something is at compared to the name for the column.

```python
# Key[Radius, Mass]
dictionary = {"earth": [6.4e6, 6e24], "mars":[3.4e6, 6.4e23], "mercury": [2.44e6, 3.3e23],
              "jupiter":[6.9e7, 1.89e27]}
dictionary['earth'][0]
```

In Python, a library refers to a collection of pre-written code or modules that can be used to perform specific tasks or functions. Libraries provide a set of functions and methods that can be utilized by developers to simplify their programming tasks and avoid reinventing the wheel. A common library used for data science is [pandas](https://pandas.pydata.org/docs/user_guide/index.html#user-guide).

```python
import pandas as pd

#make a pandas dataframe using the dictionary we made above
df = pd.DataFrame(dictionary, index = ["radius", "mass"])
```

Let's examine the data.

```python
df
```

```python
df['earth']['radius']
```

```python
#you can get the columns as a list

cols = df.columns.to_list()
```

```python
print('The columns for the dataframe are:', cols)

print('The shape of the dataframe is:', df.shape)
```

## Functions
Python functions are reusable blocks of code that perform a specific task. It allows us to organize our code into separate components. They are a way of converting complicated lists of tasks into a simple one line input. They are used very frequently by astronomers to save time and make our code cleaner. Functions in Python are defined using the ``def`` keyword, followed by the function name and parentheses containing optional input parameters. Functions can return a value or set of values, think of them as input-output machines. When a function is called, it will do everything inside its "block" and spit out whatever the return command has in it. Functions enable the division of complex tasks into smaller, manageable parts. They promote code reuse, as functions can be called multiple times with different arguments.

Notice that the description of the function is in two sets of ``'''``, this is known as a doc string. It is good practice to use these to explain your function, inputs and outputs.

```python
def example_function(x, m, b):
  '''
  This function will return a linear function of the form y = mx + b

  Inputs: x: x value, m: slope, b: y-intercept
  Outputs: y: y values computed using y = mx + b
  '''
  y = m*x + b
  return y
```

Let's try to use the functin we defined above!

```python
#define variable values
slope = 10.
y_int = -5.
x = 2.

y = example_function(x, slope, y_int)

print(y)
```

```python

y = example_function(2, 10, -5)
print(y)
```

The order in which you do things matters! Here is an example of how Python doesn't always behave the way we want it to.

```python
# Python can be tricky
number1 = 8.0
sum = number1 + number2
number2 = 10.0
print(sum)
```

If we re-define the number first, then we are able to get the correct answer. Be careful about how you define your variables!

```python
number2 = 10.0
number1 = 8.0
sum = number1 + number2
print(sum)
```

So why does Python do this? It has something to do with how Python access and stores data in the computer. When we define "number1" and "number2" there are stored in specific locations in computer memory known as "loc". When we define sum, Python asks the computer to "add the item at location 14 and the item at location 72". This is now set to be the defenition of the variable "sum". If we were to redefine "number2", then nothing changes in sum. It still accesses the two locations in memory which we previously asked it too. When we ask Python, ``print(sum)`` it simply prints those locations. Hence we get the wrong answer. If we want to print the correct answer, we must redefine sum. Then we are telling Python "no don't add the values of the numbers at location 14 and location 72, instead add the values of the numbers at location 14 and location 51. If you don't fully understand this, its okay! But this is exactly what's going on under the hood when Python does not give us the answer we want, or were expecting.
